```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from tqdm import tqdm
```

# Load raw data

```python
df = pd.read_csv('../tmp/matched_sequences.csv', header=0, index_col=0)
df.head()
```

# Embedding dev

```python
dictionary = ['B', 'E', 'D', 'G', 'I', 'H', 'L', 'O', 'P', 'S', 'R', 'T']
dict_lookup = pd.DataFrame(np.eye(len(dictionary), dtype=np.int), index=dictionary)
dict_lookup
```

```python
# Naive embedding for class 1
X1 = pd.DataFrame(columns=dict_lookup.index)
df1 = df[df['Class'] == 1]
for x, index in tqdm(zip(df1['Seq'], df1.index)):
    X1.loc[index] = np.sum([dict_lookup.loc[b, :] for b in x[::2]], axis=0)
```

```python
# Naive embedding for class 0
X0 = pd.DataFrame(columns=dict_lookup.index)
df0 = df[df['Class'] == 0]
for x, index in tqdm(zip(df0['Seq'], df0.index)):
    X0.loc[index] = np.sum([dict_lookup.loc[b, :] for b in x[::2]], axis=0)
```
